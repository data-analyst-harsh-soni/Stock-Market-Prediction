```python
import numpy as np 
import pandas as pd
```

```python
BASE_PATH = r"D:\DATA_ANALYST\FULL_STACK_FROJECT\Stock Market Prediction\stock_market_unclean_dataset"
df = pd.read_csv(BASE_PATH + r"\news_sentiment\daily_sentiment.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df.dropna(subset=['date'], inplace=True)
```

```python
df['company'] = df['company'].str.strip().str.upper()
```

```python
df = df[df['sentiment_score'].between(-1, 1)]
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['sentiment_label'] = df['sentiment_score'].apply(lambda x: 'Positive' if x > 0 else 'Negative' if x < 0 else 'Neutral')
```

```python
daily_sentiment = (df.groupby(['date', 'company'], as_index=False).agg(avg_sentiment=('sentiment_score', 'mean')))
```

```python
daily_sentiment['sentiment_strength'] = daily_sentiment['avg_sentiment'].abs()
```

```python
df.head()
```

```python
df = df.rename(columns={
    "date": "trade_date"
})
df = df[["trade_date",
         "company",
         "sentiment_score",
         "sentiment_label"]]
```

```python
df.to_csv("daily_sentiment.csv", index=False)
```
