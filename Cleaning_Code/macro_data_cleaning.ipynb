```python
import numpy as np 
import pandas as pd
```

```python
BASE_PATH = r"D:\DATA_ANALYST\FULL_STACK_FROJECT\Stock Market Prediction\stock_market_unclean_dataset"
df = pd.read_csv(BASE_PATH + r"\macro_data\inflation_interest.csv")
```

```python
df.head()
```

```python
df.rename(columns={'inflation': 'inflation_rate'}, inplace=True)
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df.dropna(subset=['date'], inplace=True)
df.sort_values(by='date', inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df = df[df['inflation_rate'] >= 0]
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
df['real_interest_rate'] = df['interest_rate'] - df['inflation_rate']
```

```python
df['inflation_trend'] = df['inflation_rate'].diff()
```

```python
df['interest_rate_trend'] = df['interest_rate'].diff()
```

```python
df['macro_environment'] = df.apply( lambda x: 'Tight' if x['real_interest_rate'] > 0 else 'Loose',axis=1 ) 
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df = df.rename(columns={
    "date": "trade_date"
})
df = df[["trade_date",
         "inflation_rate",
         "interest_rate",
         "real_interest_rate",
         "inflation_trend",
         "interest_rate_trend",
         "macro_environment"]]
```

```python
df.to_csv("inflation_interest.csv", index=False)
```
